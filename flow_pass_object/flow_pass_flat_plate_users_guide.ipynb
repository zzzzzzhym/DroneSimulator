To use flow_pass_flat_plate, users can simply define the wall parameters and can get the flow velocity at any point given a free flow velocity. The following example shows how to query the velocity of a given coordinate and to plot the flow field.

```python
import numpy as np
import matplotlib.pyplot as plt
import flow_pass_flat_plate

wall_norm = np.array([1.0, 0.0, 0.0])   # facing x-axis
wall_origin = np.array([-0.5, 0.0, 0.0])
wall_length = 8.0    
flow = flow_pass_flat_plate.FlowPassFlatPlate.Interface(wall_norm, wall_origin, wall_length)
u_free=np.array([-10.0, 10.0, 0.0])
velocity_at_one_point = flow.get_solution(u_free, np.array([1.0, 1.0, 1.0]))
print("velicty at one point")
print(velocity_at_one_point)
flow.analyze_velocity_field(u_free)
```
