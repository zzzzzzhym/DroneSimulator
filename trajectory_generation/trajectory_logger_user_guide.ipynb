This module generate a bunch of atomic paths where the vehicle takes off from hover, transfer via a trajectory and then hover again. The good thing about start and stop at hover is that the vehicle can experience a relatively even distributed velocity and acceleration. And most importantly, these paths can get connected to form continuous a random trajectories. 

```python
import matplotlib.pyplot as plt
import os
import sys

# Get the current directory of the notebook
notebook_dir = os.getcwd()

# Add the parent directory to sys.path
parent_dir = os.path.abspath(os.path.join(notebook_dir, '..'))
if parent_dir not in sys.path:
    sys.path.append(parent_dir)

import trajectory_logger

loaded_trajectory = trajectory_logger.read_trajectory("5_random_wp_map_199.pkl")
loaded_trajectory.plot_trajectory()    
plt.show()
```
